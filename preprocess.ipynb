```python
import os 
import cv2
import matplotlib.pyplot as plt
```

```python
#Read image
base = 'data/bunny'
raw = [os.path.join(os.getcwd(),base,i) for i in os.listdir(base)]
raw[:3]
```

```python
#Read
ori = []

for i in raw:
    img = cv2.imread(i,1)
#     img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ori.append(img)
```

```python
len(ori)
```

```python
plt.imshow(img)
```

```python
#Downsample
h,w,c = img.shape
img_ = cv2.resize(img,(w//2,h//2))

plt.imshow(img_)
```

```python
#Downsample
down = []
h,w,c = img.shape

for i in ori:
    img_ = cv2.resize(i,(w//2,h//2))
    down.append(img_)
```

```python
#Save out
new_base = 'data/bunny_360_640'

if not os.path.isdir(new_base):
    os.makedirs(new_base)
    

for idx,i in enumerate(down):
    
    name = '%04d.png' % idx
    tgt = os.path.join(new_base,name)
    
    cv2.imwrite(tgt,i)
```
