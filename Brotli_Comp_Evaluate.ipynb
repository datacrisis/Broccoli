```python
#Generate raw commands to be run
# raw = '/scratch/kl3866/IVP/HNeRV/batches/output/2x_8bit_gamut/1_1_1__Dim64_16_FC9_16_KS0_1_5_RED1.2_low12_blk1_1_e600_b1_quant_M8_E6_lr0.001_cosine_0.1_1_0.1_L2_Size1/quant_vid.pth'
# comp = '/scratch/kl3866/IVP/HNeRV/batches/final/2x_1080_HNeRV_gamut/brotli/2x_REPLACE_QTM32_QTE6/compressed_quant_vid.pth.br'

compped = []

for i in seqs:
    mid = os.listdir(i)[0]
    src = os.path.join(i,mid,'quant_vid.pth')
    valid = os.path.isfile(src) #checker
    assert valid, Exception("Warning, src file does not exist!")
   
    tgt = os.path.join(target_dir, i.split('/')[-1], 'compressed_quant_vid.pth.br')
    valid = os.path.isdir('/'.join(tgt.split('/')[:-1])) #checker; remove file name
    assert valid, Exception("Warning, src file does not exist!")
    
    cmd = 'brotli -q 11 -o {} {}'.format(tgt,src)
    compped.append(tgt) #store for later use
#     print("===="*10)
    print('\n')
    print(cmd)
```

```python
ct = {'Twilight': 600,
 'SunBath': 300,
 'Bosphorus': 600,
 'RaceNight': 600,
 'FlowerKids': 600,
 'Beauty': 600,
 'CityAlley': 600,
 'ReadySteadyGo': 600,
 'FlowerFocus': 600,
 'Jockey': 600,
 'YatchRide': 600,
 'HoneyBee': 600,
 'ShakeNDry': 300,
 'RiverBank': 600}
```

```python
#Compute BPP
bpps = {}

#Params
res = 1920 * 960 #res for 2x output

for f in compped:
    
    #Inclusive of embed and decoder already
    vid = f.split('/')[-2].split('_')[1]
    nframes = ct[vid]
    
    total_bits = os.path.getsize(f) * 8 #bytes -> bits
    total_bpp = total_bits / (res * nframes)
    
    
    #Store
    bpps[vid] = {'bpp': total_bpp,
                 'path': f,
                 'res':'1920_960',
                 'nframes':nframes}
```

```python
import pandas as pd
df = pd.DataFrame(bpps).T
df
```

```python
df.to_csv('results/HNeRV_2x_1920x960_8bit_brotli_BPP.csv') #save
```
