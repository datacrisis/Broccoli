### HEVC Plot

```python
df_hevc = pd.read_csv('hevc_yuv_2.csv')
```

```python
df_hevc.columns
```

```python
X_inter = []
Y_inter = []
QP_list = [17, 22, 27, 32, 37, 42]
for i in range(len(QP_list)):
    X_inter.append(df_hevc[(df_hevc['QP']==QP_list[i])&(df_hevc['Inter']==True)].BPP.mean())
    Y_inter.append(df_hevc[(df_hevc['QP']==QP_list[i])&(df_hevc['Inter']==True)].PSNR.mean())

X_inter = X_inter[::-1]
Y_inter = Y_inter[::-1]
plt.plot(X_inter,Y_inter)

X_intra = []
Y_intra = []
QP_list = [17, 22, 27, 32, 37, 42]
for i in range(len(QP_list)):
    X_intra.append(df_hevc[(df_hevc['QP']==QP_list[i])&(df_hevc['Intra']==True)].BPP.mean())
    Y_intra.append(df_hevc[(df_hevc['QP']==QP_list[i])&(df_hevc['Intra']==True)].PSNR.mean())

X_intra = X_intra[::-1]
Y_intra = Y_intra[::-1]
plt.plot(X_intra,Y_intra)

plt.legend(['Inter Coding', 'Intra coding'])
```
